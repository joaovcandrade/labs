# 1a parte

No espaço de código abaixo, plote as 3 imagens, uma ao lado da outra. Ainda, plote o histograma de cada canal de cor (R - Vermelho, G - Verde e B - Azul) abaixo da respectiva imagem.

```
imagem_A = Image.open('/content/drive/My Drive/Colab Notebooks/lab02/bailey-zindel-NRQV-hBF10M-unsplash.jpg')
imagem_B = Image.open('/content/drive/My Drive/Colab Notebooks/lab02/claudio-testa--SO3JtE3gZo-unsplash.jpg')
imagem_C = Image.open('/content/drive/My Drive/Colab Notebooks/lab02/johannes-plenio-RwHv7LgeC7s-unsplash.jpg')

plt.figure(figsize=(20,4))

plt.subplot(131) 
plt.title('Imagem A')
plt.axis('off')
plt.imshow(imagem_A)

plt.subplot(132) 
plt.title('Imagem B')
plt.axis('off')
plt.axis('off')
plt.imshow(imagem_B)

plt.subplot(133) 
plt.title('Imagem C')
plt.axis('off')
plt.imshow(imagem_C)

plt.show()

img_arr_A = np.array(imagem_A)
red_A= img_arr_A[:,:,0]
green_A= img_arr_A[:,:,1]
blue_A= img_arr_A[:,:,2]

img_arr_B = np.array(imagem_B)
red_B= img_arr_B[:,:,0]
green_B= img_arr_B[:,:,1]
blue_B= img_arr_B[:,:,2]

img_arr_C = np.array(imagem_C)
red_C= img_arr_C[:,:,0]
green_C= img_arr_C[:,:,1]
blue_C= img_arr_C[:,:,2]

#----------------------------------#
plt.figure(figsize=(12,12))

plt.subplot(331) 
plt.hist(red_A.ravel(), color='red')

plt.subplot(334) 
plt.hist(green_A.ravel(), color='green')

plt.subplot(337) 
plt.hist(blue_A.ravel(), color='blue')

plt.subplot(332) 
plt.hist(red_B.ravel(), color='red')

plt.subplot(335) 
plt.hist(green_B.ravel(), color='green')

plt.subplot(338) 
plt.hist(blue_B.ravel(), color='blue')

plt.subplot(333) 
plt.hist(red_C.ravel(), color='red')

plt.subplot(336) 
plt.hist(green_C.ravel(), color='green')

plt.subplot(339) 
plt.hist(blue_C.ravel(), color='blue')

plt.show()
```

Olhando para os gráficos gerados, discorram comparando os resultados.

>Na imagem A percebe-se maior predominância de tons azuis e verdes. Tons vermelhos médios estão pouco presente.

>Na imagem B percebe-se maior predominância de tons verdes devido a maior escala e tons fracos de azul

>Na imagem C percebe-se pouca presença do azul, sendo em sua maioria tons fracos, grande predominância de diversos tons vermelhos e prença de tons verde fraco.






# 2a parte

Escolha uma das imagens acima, para continuarmos os experimentos. 

Vamos gerar algumas imagens preto e branco, utilizando limiares diferentes:

*  45
*  100
*  145
*  200
*  245

Escreva o código no espaço abaixo.

```
limiares = [45, 100, 145, 200, 245]
imagens = []

img = np.mean(img_arr_A, axis=2)
for l in limiares:
  img_cp = img.copy()
  img_cp[img_cp >= l] = 255;
  img_cp[img_cp < l] = 0;

  imagens.append(img_cp)
```

Plote as imagens geradas. (Escreva o código no espaço abaixo)

```
plt.figure(figsize=(40,40))

plt.subplot(3,3,1) 
plt.imshow(Image.fromarray(img))

for i,im in enumerate(imagens):
  plt.subplot(3,3,int(i)+2) 
  plt.imshow(Image.fromarray(im))

plt.show()
```

O que você consegue extrair deste experimento? Escreva no espaço abaixo.



> Quando mais baixo o limiar, a imagem tende a obter mais pontos brancos. Quanto mais alto o limiar, a imagem tende a obter pontos pretos.
